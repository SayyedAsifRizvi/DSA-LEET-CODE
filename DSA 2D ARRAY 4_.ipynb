```python
CODE:
class Solution:
    def findDifference(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:

        # define set to store elements
        s1 = set(nums1)
        s2 = set(nums2)

        # define output
        out = [[],[]]

        # adding the elements to output if not contains in the set
        for i in s1:
            if i not in s2:  
                out[0].append(i)
        for i in s2:
            if i not in s1:
                out[1].append(i)
        
        return out
LINK:https://leetcode.com/problems/find-the-difference-of-two-arrays/description/
```

```python
CODE:
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        
        # Two Pointer Approach
        n = len(nums)
        l, r = 0, n - 1
        k = n - 1
        ans = [0] * n
        while k >= 0:
            if abs(nums[l]) > nums[r]:
                ans[k] = nums[l] * nums[l]
                l += 1
            else:
                ans[k] = nums[r] * nums[r]
                r -= 1
            k -= 1
        return ans
LINK:https://leetcode.com/problems/squares-of-a-sorted-array/description/
```

```python
CODE:
class Solution:
    def maxCount(self, m: int, n: int, ops: List[List[int]]) -> int:
        min_row = m
        min_col = n
        for i in range(len(ops)):
            min_row=min(min_row, ops[i][0])
            min_col=min(min_col, ops[i][1])
        return min_row*min_col
LINK:https://leetcode.com/problems/range-addition-ii/description/
```
